```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("Algerian_forest_fires_cleaned.csv")
```

```python
df.head()
```

```python
df.columns
```

```python
df.drop(["day","month","year"],axis=1,inplace=True)
```

fix spacing

```python
df["Classes"]=np.where(df['Classes'].str.contains("not fire"),0,1)
```

```python
df['Classes'].value_counts()
```

Divide into independent and dependent features

```python
X=df.drop('FWI',axis=1)
```

```python
y=df["FWI"]
```

```python
X.head()
```

```python
y.head()
```

Train test split

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=42)
```

```python
X_train.shape,X_test.shape
```

Feature Selection based on correlation

```python
X_train.corr()
```

Check for multicollinearity

```python
plt.figure(figsize=(10,8))
corr=X_train.corr()
sns.heatmap(corr,cmap='coolwarm',annot=True)
```

```python
corr.columns
```

Note keeping independent features with very high correlation just means you are passing multiple data to the points where data learns from all and try to give the best line which might cause overfitting so basically you have to keep the most important feature while removing other highly correlated features.
But here we will delete all features exceeding correlation of 0.85.

```python
def correlation(dataset,threshold):
    col_corr=set()
    corr_matrix=dataset.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i,j])>threshold:
                col_name=corr_matrix.columns[i]
                col_corr.add(col_name)

    return col_corr
```

Gives all columns which have correlation greater than 0.85

```python
corr_features=correlation(X_train,0.85)
```

Dropping these featureswhen correlation is greater than 0.85

```python
X_train.drop(corr_features,axis=1,inplace=True)
X_test.drop(corr_features,axis=1,inplace=True)
X_train.shape,X_test.shape
```

Feature Scaling or Standarization

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X_train_scaled=scaler.fit_transform(X_train)
X_test_scaled=scaler.transform(X_test)
```

```python
X_train_scaled
```

Box plots to understand the effect of Standard Scaler

```python
plt.subplots(figsize=(15,10))
plt.subplot(1,2,1)
sns.boxplot(data=X_train)
plt.title('X_train Before Scaling')

plt.subplot(1,2,2)
sns.boxplot(data=X_train_scaled)
plt.title('X_train After Scaling')
plt.show()
```

Linear Regression Model

```python
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
reg.fit(X_train_scaled,y_train)

from sklearn.metrics import mean_absolute_error,r2_score

y_pred=reg.predict(X_test_scaled)
mae=mean_absolute_error(y_test,y_pred)
r2=r2_score(y_test,y_pred)
print(mae)
print(r2)
plt.scatter(y_test,y_pred)
```

Since y_test and y_pred are following linear relation so we can say that our model is good
