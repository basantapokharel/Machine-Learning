```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
from sklearn.datasets import fetch_openml
```

```python
# Fetch the MNIST dataset
mnist = fetch_openml('mnist_784', version=1)
```

```python
mnist
```

```python
X, y = mnist["data"], mnist["target"]
```

```python
# Convert the labels to integers
y = y.astype(np.int32)


# Filter out only the "Zero" digit (label = 0)
X_zero = X[y == 0]

# Select the first image of the digit "Zero" and reshape it into a 2D array (28x28) using iloc
sample_image = X_zero.iloc[0].values.reshape(28, 28)

# Use seaborn's heatmap to visualize the pixel values
plt.figure(figsize=(6, 6))
sns.heatmap(sample_image, cmap='Greys', cbar=False, square=True, xticklabels=False, yticklabels=False)

# Add a title and show the plot
plt.title("Visualization of the Digit 'Zero' using Seaborn")
plt.show()
```

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier()
rf.fit(X,y)
```

```python
rf.feature_importances_
```

```python
rf.feature_importances_.shape
```

```python
sns.heatmap(rf.feature_importances_.reshape(28,28))
```
