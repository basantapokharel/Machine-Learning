```python
import numpy as np
import pandas as pd
```

```python
from sklearn.datasets import make_classification
from sklearn.tree import DecisionTreeClassifier
```

```python
X,y=make_classification(n_samples=20,n_classes=2,n_features=2,n_informative=2,n_redundant=0,random_state=32)
```

```python
clf=DecisionTreeClassifier()
clf.fit(X,y)
```

```python
from sklearn.tree import plot_tree
plot_tree(clf)
```

```python
clf.feature_importances_
```

Total sum is 1 since it is normalized

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf=RandomForestClassifier(n_estimators=2)
rf.fit(X,y)
```

```python
rf.feature_importances_
```

```python
rf.estimators_
```

```python
rf.estimators_[0].feature_importances_
```

```python
rf.estimators_[1].feature_importances_
```
