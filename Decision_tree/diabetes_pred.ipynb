```python
import numpy as np
import pandas as pd
```

```python
from sklearn.datasets import load_diabetes
```

```python
data=load_diabetes()
```

```python
print(data['DESCR'])
```

```python
data
```

```python
df=pd.DataFrame(data.data,columns=['age','sex','bmi','bp','s1','s2','s3','s4','s5','s6'])
```

```python
X=df
y=data['target']
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=40)
```

```python
X_train.corr()
```

```python
from sklearn.tree import DecisionTreeRegressor
reg=DecisionTreeRegressor()
reg.fit(X_train,y_train)
```

Hyperparameter Tuning

```python
param = {
    'criterion': ["squared_error", "friedman_mse", "absolute_error", "poisson"],
    'splitter': ["best", "random"],
    'max_depth': [1, 2, 3, 7, 8, 10, 13, 15, 17, 18, 20, 23, 26],
    'max_features': ['auto', "sqrt", "log2"]
}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
regre=DecisionTreeRegressor()
```

```python
grid=GridSearchCV(regre,param_grid=param,cv=5,scoring='neg_mean_squared_error')
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
y_pred=grid.predict(X_test)
```

```python
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
```

```python
print(r2_score(y_test,y_pred))
print(mean_absolute_error(y_test,y_pred))
print(mean_squared_error(y_test,y_pred))
```

TO see decision tree

```python
model = DecisionTreeRegressor(criterion='squared_error', 
                              max_depth=3, 
                              max_features='sqrt', 
                              splitter='best')
```

```python
model.fit(X_train,y_train)
```

Visualize decision tree

```python
import matplotlib.pyplot as plt
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(model,filled=True)
```
