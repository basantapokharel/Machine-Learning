```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('hierarchical-clustering-with-python-and-scikit-learn-shopping-data.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
df=df.drop(columns=['CustomerID','Genre','Age'])
```

```python
df.head()
```

```python
data=df.values
```

```python
data
```

```python
import scipy.cluster.hierarchy as shc
import matplotlib.pyplot as plt

# Assuming `data` is defined and contains your dataset
plt.figure(figsize=(10, 7))
plt.title('Customer Dendrogram')
dend = shc.dendrogram(shc.linkage(data, method='ward'))
plt.show()
```

so 5 clusters 

```python
from sklearn.cluster import AgglomerativeClustering

# Create an AgglomerativeClustering model with 5 clusters and 'ward' linkage
cluster = AgglomerativeClustering(n_clusters=5, linkage='ward')

# Fit the model to your data and predict cluster labels
labels = cluster.fit_predict(data)
```

```python
labels
```

```python
plt.figure(figsize=(10,7))
plt.scatter(data[:,0],data[:,1],c=cluster.labels_,cmap='rainbow')
```
